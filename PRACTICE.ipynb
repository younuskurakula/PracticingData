# B Clean The Data, Detecting Missing,Duplicates and Filling the Data

# df.isnull(),.isnull().sum(),.isnull().sum().sum()......  # Shows Nan Or None Values

```python
df1=df.isnull() # isnull() show True means NaN or None values and False valid Values
df1
```

```python
df11=df.isnull().sum() # isnull().sum() NaN or None values column wise
df11
```

```python
df111=df.isnull().sum().sum() # Whole table NaN or None Values you have
df111
```

```python
# df=df.dropna(),.dropna(subset=['']),.dropna(axis=1) # Drops the Nan or None Values
```

```python
df2=df.dropna() # Drops all NaN or None values , shows only valid row values
df2
```

```python
df22=df.dropna(subset=['Age']) # drop N,None values in Age Column it shows only valid values in Age
df22
```

```python
df222=df.dropna(subset=['Salary','Score','Grade']) # drops Nan,None values in Salary,Score,Grade.It shows only valid values of Salar,Score,Grade
df222
```

```python
df2222=df.dropna(axis=1)
df2222
```

```python
df
```

# df=df.fillna(), .fillna(df[''].mean()), .fillna(df[''].max())........ # Filling the Data with necessary inputs

```python
df3=df.fillna(56) 
df3
```

```python
df33=df.fillna(df['Age'].mean())
df33
```

```python
# Shortcut for filling all columns at a time
df333=df.fillna({
    'Age':df['Age'].mean(),
    'Score':df['Score'].max(),
    'Salary':df['Salary'].std(),
    'Grade':df['Grade'].ffill()
})

df333
```

# Filtering Data with &,|,~,.isin,.between,.str.contains,.query

```python
df['Age']>30
```

```python
df[(df['Age']>30)&(df['Score']>50)]
```

```python
df[~(df['Name']=='Saddam')]
```

```python
df[(df['Age']>30) | (df['Salary']>10000)]
```

```python
df[df['Name'].isin(['Yasmin','Enus'])]
```

```python
df[df['Age'].between(20,30)]
```

```python
df[df['Name'].str.contains("m")]
```

```python
df.query('Age>30')
```
